We are trying to see if there is a relationship between failed / passed course and other variables. For this we are going encode the final grade (G3) as failed, -1, (grades below 10) and passed, 1, (grades above 10). The classification shall be accomplished via decision trees.

```python
import numpy as np
import pandas as pd
import graphviz
from sklearn import tree

students_portuguese = pd.read_csv('../data/student-por.csv', sep = ';')

students_portuguese
```

```python
students_portuguese['paid']       = students_portuguese['paid'].apply(lambda x: -1 if x == "no" else 1)
students_portuguese['address']    = students_portuguese['address'].apply(lambda x: -1 if x == "R" else 1)
students_portuguese['famsize']    = students_portuguese['famsize'].apply(lambda x: -1 if x == "GT3" else 1)
students_portuguese['Pstatus']    = students_portuguese['Pstatus'].apply(lambda x: -1 if x == "A" else 1)
students_portuguese['Mjob']       = students_portuguese['Mjob'].apply(lambda x: 1 if x == "teacher" else 2 \
                                                                  if x == "health" else 3 if x == "services" else 4 \
                                                                  if x == "at_home" else 5)
students_portuguese['Fjob']       = students_portuguese['Fjob'].apply(lambda x: 1 if x == "teacher" else 2 \
                                                                  if x == "health" else 3 if x == "services" else 4 \
                                                                  if x == "at_home" else 5)
students_portuguese['reason']     = students_portuguese['reason'].apply(lambda x: 1 if x == "home" else 2 \
                                                                    if x == "reputation" else 3 if "course" else 4)
students_portuguese['guardian']   = students_portuguese['guardian'].apply(lambda x: 1 if x == "mother" else 2 \
                                                                      if x == "father" else 3)
students_portuguese['schoolsup']  = students_portuguese['schoolsup'].apply(lambda x: -1 if x == "no" else 1)
students_portuguese['famsup']     = students_portuguese['famsup'].apply(lambda x: -1 if x == "no" else 1)
students_portuguese['activities'] = students_portuguese['activities'].apply(lambda x: -1 if x == "no" else 1)
students_portuguese['nursery']    = students_portuguese['nursery'].apply(lambda x: -1 if x == "no" else 1)
students_portuguese['higher']     = students_portuguese['higher'].apply(lambda x: -1 if x == "no" else 1)
students_portuguese['internet']   = students_portuguese['internet'].apply(lambda x: -1 if x == "no" else 1)
students_portuguese['romantic']   = students_portuguese['romantic'].apply(lambda x: -1 if x == "no" else 1)

students_portuguese['G1'] = students_portuguese['G1'].apply(lambda x: -1 if x < 10 else 1)
students_portuguese['G2'] = students_portuguese['G2'].apply(lambda x: -1 if x < 10 else 1)
students_portuguese['G3'] = students_portuguese['G3'].apply(lambda x: -1 if x < 10 else 1)
```

```python
students_portuguese
```

```python
features = students_portuguese.loc[:, ['address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason', 'guardian', 'schoolsup', 
                                    'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic',
                                    'Medu', 'Fedu', 'studytime', 'failures', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc',
                                    'health', 'G2', 'G1', 'absences', 'age', 'traveltime']]

target   = students_portuguese.loc[:, 'G3']
```

```python
tree_full = tree.DecisionTreeClassifier()
tree_full = tree_full.fit(features, target)
```

```python
data_tree_full  = tree.export_graphviz(tree_full, out_file = None,
                                      feature_names = features.columns, filled = True, rounded = True,
                                      special_characters = True, class_names = ["Fail", "Pass"])
graph_tree_full = graphviz.Source(data_tree_full)
graph_tree_full
```

The above is not the most easily interpretable tree, and very large so we are going to turn it down a little ...

```python
features = students_portuguese.loc[:, ['famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason', 'guardian', 'schoolsup', 
                                    'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic',
                                    'Medu', 'Fedu', 'studytime', 'failures', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc',
                                    'health']]
```

```python
tree_full = tree.DecisionTreeClassifier()
tree_full = tree_full.fit(features, target)

data_tree_full  = tree.export_graphviz(tree_full, out_file = None,
                                      feature_names = features.columns, filled = True, rounded = True,
                                      special_characters = True,
                                      class_names = ["Fail", "Pass"])
graph_tree_full = graphviz.Source(data_tree_full)
graph_tree_full
```

The above is somehow better, we reduce the max depth to see if it's more easily interpretable:

```python
data_tree_full  = tree.export_graphviz(tree_full, out_file = None,
                                      feature_names = features.columns, filled = True, rounded = True,
                                      special_characters = True, max_depth = 5,
                                      class_names = ["Fail", "Pass"])
graph_tree_full = graphviz.Source(data_tree_full)
graph_tree_full
```

```python
features = students_portuguese.loc[:, ['Pstatus', 'famsup', 'higher', 'Medu', 'Fedu', 'failures', 'famrel']]

tree_simple = tree.DecisionTreeClassifier()
tree_simple = tree_simple.fit(features, target)

data_tree  = tree.export_graphviz(tree_simple, out_file = None,
                                      feature_names = features.columns, filled = True, rounded = True,
                                      special_characters = True,
                                      class_names = ["Fail", "Pass"])
graph_tree_simple = graphviz.Source(data_tree)
graph_tree_simple
```

```python
data_tree  = tree.export_graphviz(tree_simple, out_file = 'portuguese_simplified.dot',
                                      feature_names = features.columns, filled = True, rounded = True,
                                      special_characters = True,
                                      class_names = ["Fail", "Pass"])
```

The previous tree simply states that if the student is interested in pursing higher education than he is more likely to pass, so we are going to remove that and see what happens.

```python
features = students_portuguese.loc[:, ['Pstatus', 'famsup', 'Medu', 'Fedu', 'failures', 'famrel']]

tree_simple = tree.DecisionTreeClassifier()
tree_simple = tree_simple.fit(features, target)

data_tree  = tree.export_graphviz(tree_simple, out_file = None,
                                      feature_names = features.columns, filled = True, rounded = True,
                                      special_characters = True,
                                      class_names = ["Fail", "Pass"])
graph_tree_simple = graphviz.Source(data_tree)
graph_tree_simple
```

```python
data_tree  = tree.export_graphviz(tree_simple, out_file = 'portuguese_simplified_without_higher.dot',
                                      feature_names = features.columns, filled = True, rounded = True,
                                      special_characters = True,
                                      class_names = ["Fail", "Pass"])
```

```python
features = students_portuguese.loc[:, ['Pstatus', 'famsup', 'Medu', 'Fedu', 'famrel']]

tree_simple = tree.DecisionTreeClassifier()
tree_simple = tree_simple.fit(features, target)

data_tree  = tree.export_graphviz(tree_simple, out_file = None,
                                      feature_names = features.columns, filled = True, rounded = True,
                                      special_characters = True,
                                      class_names = ["Fail", "Pass"])
graph_tree_simple = graphviz.Source(data_tree)

data_tree  = tree.export_graphviz(tree_simple, out_file = 'portuguese_failures_free.dot',
                                      feature_names = features.columns, filled = True, rounded = True,
                                      special_characters = True,
                                      class_names = ["Fail", "Pass"])

graph_tree_simple
```

The interesting aspects from this one: if the education of the father is quite high or the family relations are quite good than the student is quite likely to pass.
