```python
students_combined_outer = pd.merge(students_portuguese, students_mathematics, on = key, how = "outer")

students_combined_outer
```

```python
students_combined_outer.columns
```

```python
topics = {'school': [], 'sex': [], 'age': [], 'address': [], 'famsize': [], 'Pstatus': [], 'Medu': [],
          'Fedu': [], 'Mjob': [], 'Fjob': [], 'reason': [], 'guardian': [], 'traveltime': [],
          'studytime': [], 'failures': [], 'schoolsup': [], 'famsup': [], 'paid_x': [],
          'activities': [], 'nursery': [], 'higher': [], 'internet': [], 'romantic': [],
          'famrel': [], 'freetime': [], 'goout': [], 'Dalc': [], 'Walc': [], 'health': [],
          'absences_x': [], 'G1_x': [], 'G2_x': [], 'G3_x': [], 'paid_y': [], 'absences_y': [],
          'G1_y': [], 'G2_y': [], 'G3_y': [], 'classes': []}

for iter in students_combined_outer.index:
    
    key = 0
    if np.isnan(students_combined_outer['absences_x'][iter]) == True:
        key = key + 1
    if np.isnan(students_combined_outer['absences_y'][iter]) == True:
        key = key + 2
    
    topics['school'].append(students_combined_outer['school'][iter])
    topics['sex'].append(students_combined_outer['sex'][iter])
    topics['age'].append(students_combined_outer['age'][iter])
    topics['address'].append(students_combined_outer['address'][iter])
    topics['famsize'].append(students_combined_outer['famsize'][iter])
    topics['Pstatus'].append(students_combined_outer['Pstatus'][iter])
    topics['Medu'].append(students_combined_outer['Medu'][iter])
    topics['Fedu'].append(students_combined_outer['Fedu'][iter])
    topics['Mjob'].append(students_combined_outer['Mjob'][iter])
    topics['Fjob'].append(students_combined_outer['Fjob'][iter])
    topics['reason'].append(students_combined_outer['reason'][iter])
    topics['guardian'].append(students_combined_outer['guardian'][iter])
    topics['traveltime'].append(students_combined_outer['traveltime'][iter])
    topics['studytime'].append(students_combined_outer['studytime'][iter])
    topics['failures'].append(students_combined_outer['failures'][iter])
    topics['schoolsup'].append(students_combined_outer['schoolsup'][iter])
    topics['famsup'].append(students_combined_outer['famsup'][iter])
    topics['paid_x'].append(students_combined_outer['paid_x'][iter])
    topics['activities'].append(students_combined_outer['activities'][iter])    
    topics['nursery'].append(students_combined_outer['nursery'][iter])
    topics['higher'].append(students_combined_outer['higher'][iter])
    topics['internet'].append(students_combined_outer['internet'][iter])
    topics['romantic'].append(students_combined_outer['romantic'][iter])
    topics['famrel'].append(students_combined_outer['famrel'][iter])
    topics['freetime'].append(students_combined_outer['freetime'][iter])
    topics['goout'].append(students_combined_outer['goout'][iter])
    topics['Dalc'].append(students_combined_outer['Dalc'][iter])
    topics['Walc'].append(students_combined_outer['Walc'][iter])
    topics['health'].append(students_combined_outer['health'][iter])
    topics['absences_x'].append(students_combined_outer['absences_x'][iter])
    topics['G1_x'].append(students_combined_outer['G1_x'][iter])
    topics['G2_x'].append(students_combined_outer['G2_x'][iter])
    topics['G3_x'].append(students_combined_outer['G3_x'][iter])
    topics['paid_y'].append(students_combined_outer['paid_y'][iter])
    topics['absences_y'].append(students_combined_outer['absences_y'][iter])
    topics['G1_y'].append(students_combined_outer['G1_y'][iter])
    topics['G2_y'].append(students_combined_outer['G2_y'][iter])
    topics['G3_y'].append(students_combined_outer['G3_y'][iter])
    topics['classes'].append(key)
```

```python
r_topic = pd.DataFrame(topics)

r_topic
```

```python
students_combined.to_csv('../data/students-combined.csv', sep = ';', encoding = 'utf-8-sig')
r_topic.to_csv('../data/students-combined-outer.csv', sep = ';', encoding = 'utf-8-sig')
```
